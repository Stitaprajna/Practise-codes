```python
import requests
import pandas as pd
import json
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
class NseIndia:
    def __init__(self):
        self.headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Apple'
                        'WebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.134 Safari/537.36'}
        self.session = requests.Session()
        self.session.get('https://www.nseindia.com', headers= self.headers)
        
        
    def Data(self):
        data = self.session.get('https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY', headers= self.headers)
        return data
nse = NseIndia()
l=nse.Data()
```

```python
d = l.json()
```

```python
lst = []
for i in d['filtered']['data']:
  lst.append(i['CE'])
```

```python
df4 = pd.DataFrame(lst)
lst1 = []
for i in df4.columns:
  lst1.append('call_'+i)
```

```python
df4.columns = lst1
```

```python
df4.head()
```

```python
df4.rename(columns = {'call_expiryDate':'expiryDate','call_strikePrice':'strikePrice','call_underlying':'underlying','call_identifier':'identifier'}, inplace=True)
```

```python
df4.head()
```

```python
LST = []
for i in d['filtered']['data']:
  LST.append(i['PE'])
```

```python
df5 = pd.DataFrame(LST)
```

```python
df5.head()
```

```python
lst2 = []
for i in df5.columns:
  lst2.append('put_' + i)
```

```python
df5.columns = lst2
```

```python
df5.drop(['put_expiryDate','put_strikePrice','put_underlying','put_identifier'],axis=1,inplace=True)
```

```python
df5
```

```python
DF = pd.concat([df4,df5],axis=1)
```
